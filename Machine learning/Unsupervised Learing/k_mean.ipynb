```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x , y = make_blobs(n_samples=500, n_features=2, centers=5, cluster_std=1.5, random_state=42)
```

```python
x.shape
```

```python
kmean = KMeans(n_clusters=4, init='random', n_init=10, max_iter=300, random_state=42)
kmean.fit(x).score(x)
```

```python
kmean.cluster_centers_
```

```python
kmean.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=kmean.labels_, cmap='viridis', edgecolors='k')
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1], c='red', s=200, alpha=1,)
```
