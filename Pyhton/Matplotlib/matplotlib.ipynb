###  Matplotlib Pie Charts

```python
y = np.array([1,2,3,4])
plt.pie(y)
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
plt.pie(y, labels=mylabal)
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
plt.pie(y, labels=mylabal,startangle=90)
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
myexplode = [0.2,0,0,0]
plt.pie(y, labels=mylabal,startangle=90,explode=myexplode)
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
myexplode = [0.2,0,0,0]
plt.pie(y, labels=mylabal,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
myexplode = [0.2,0,0,0]
mycolor = [ "red","green","blue","yellow"]
plt.pie(y, labels=mylabal,startangle=90,explode=myexplode,shadow=True,colors=mycolor)
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
myexplode = [0.2,0,0,0]
mycolor = [ "red","green","blue","yellow"]
plt.pie(y, labels=mylabal,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.show()
```

```python
y = np.array([1,2,3,4])
mylabal = ["apple","mango","banana","orange"]
myexplode = [0.2,0,0,0]
mycolor = [ "red","green","blue","yellow"]
plt.pie(y, labels=mylabal,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.legend(title = "color ")
plt.show()
```
